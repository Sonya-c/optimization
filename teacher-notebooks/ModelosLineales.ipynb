```python
import pandas as pd
```

```python
df_iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)
```

```python
df_iris.columns = ['sepal length','sepal width', 'petal length','petal width','class']
```

```python
df_iris
```

***Analisis Visual***

```python
import seaborn as sns
```

```python
sns.scatterplot(data=df_iris, x='petal length', y='petal width',hue='class')
```

```python
sns.regplot(data=df_iris, x='petal length', y='petal width')
```

```python
sns.pairplot(data=df_iris, kind='reg')
```
